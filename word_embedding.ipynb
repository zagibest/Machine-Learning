```python
import numpy as np

def cosine_similarity(vec1, vec2):
    dot_product = np.dot(vec1, vec2)
    magnitude = np.sqrt(np.sum(vec1**2)) * np.sqrt(np.sum(vec2**2))
    similarity = dot_product / magnitude
    return similarity

word1 = np.array([10, 82, 0, 4, 275])
word2 = np.array([85, 5, 4, 0, 8])
word3 = np.array([237, 20, 4, 1, 9])

context1 = np.array([10, 85, 237])
context2 = np.array([82, 5, 20])
context3 = np.array([0, 4, 4])
context4 = np.array([4, 0, 1])
context5 = np.array([275, 8, 9])

sim1_2 = cosine_similarity(word1, word2)
sim1_3 = cosine_similarity(word1, word3)
sim2_3 = cosine_similarity(word2, word3)


print("word1 and word2:", sim1_2)
print("word1 and word3:", sim1_3)
print("word2 and word3:", sim2_3)

print("context1 and context2:", cosine_similarity(context1, context2))
print("context1 and context3:", cosine_similarity(context1, context3))
print("context1 and context4:", cosine_similarity(context1, context4))
print("context1 and context5:", cosine_similarity(context1, context5))
print("context2 and context3:", cosine_similarity(context2, context3))
print("context2 and context4:", cosine_similarity(context2, context4))
print("context2 and context5:", cosine_similarity(context2, context5))
print("context3 and context4:", cosine_similarity(context3, context4))
print("context3 and context5:", cosine_similarity(context3, context5))
print("context4 and context5:", cosine_similarity(context4, context5))

```
